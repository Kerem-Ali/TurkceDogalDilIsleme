```python
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from transformers import BertTokenizer, BertModel

# Veri seti sınıfı
class CustomDataset(Dataset):
    def __init__(self, sentences, keywords, labels, tokenizer, max_length):
        self.sentences = sentences
        self.keywords = keywords
        self.labels = labels
        self.tokenizer = tokenizer
        self.max_length = max_length

    def __len__(self):
        return len(self.sentences)

    def __getitem__(self, idx):
        sentence = self.sentences[idx]
        keyword = self.keywords[idx]
        label = self.labels[idx]

        encoding = self.tokenizer.encode_plus(
            sentence,
            keyword,
            add_special_tokens=True,
            max_length=self.max_length,
            return_token_type_ids=False,
            padding='max_length',
            truncation=True,
            return_attention_mask=True,
            return_tensors='pt'
        )

        return {
            'input_ids': encoding['input_ids'].flatten(),
            'attention_mask': encoding['attention_mask'].flatten(),
            'labels': torch.tensor(label, dtype=torch.long)
        }

# CNN model
class TextCNN(nn.Module):
    def __init__(self, num_classes, vocab_size, embedding_dim, num_filters, filter_sizes):
        super(TextCNN, self).__init__()
        self.embedding = nn.Embedding(vocab_size, embedding_dim)
        self.convs = nn.ModuleList([
            nn.Conv2d(1, num_filters, (k, embedding_dim)) for k in filter_sizes
        ])
        self.dropout = nn.Dropout(0.5)
        self.fc = nn.Linear(num_filters * len(filter_sizes), num_classes)

    def forward(self, x):
        x = self.embedding(x)
        x = x.unsqueeze(1)
        x = [nn.functional.relu(conv(x)).squeeze(3) for conv in self.convs]
        x = [nn.functional.max_pool1d(i, i.size(2)).squeeze(2) for i in x]
        x = torch.cat(x, 1)
        x = self.dropout(x)
        return self.fc(x)

# Veri yükleme ve ön işleme
# Örnek veri


sentences = [item['sentence'] for item in data]
keywords = [item['keyword'] for item in data]
labels = [item['label'] for item in data]

# Label encoding
label_encoder = LabelEncoder()
encoded_labels = label_encoder.fit_transform(labels)

# Veri bölme
train_sentences, val_sentences, train_keywords, val_keywords, train_labels, val_labels = train_test_split(
    sentences, keywords, encoded_labels, test_size=0.2, random_state=42
)

# BERT tokenizer
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
max_length = 128

# Veri yükleyicileri
train_dataset = CustomDataset(train_sentences, train_keywords, train_labels, tokenizer, max_length)
val_dataset = CustomDataset(val_sentences, val_keywords, val_labels, tokenizer, max_length)

train_loader = DataLoader(train_dataset, batch_size=1024, shuffle=True)
val_loader = DataLoader(val_dataset, batch_size=1024)

# Model parametreleri
num_classes = len(label_encoder.classes_)
vocab_size = tokenizer.vocab_size
embedding_dim = 300
num_filters = 100
filter_sizes = [3, 4, 5]

# Model, kayıp fonksiyonu ve optimizer
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
model = TextCNN(num_classes, vocab_size, embedding_dim, num_filters, filter_sizes).to(device)
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters())

# Eğitim döngüsü
num_epochs = 10000

for epoch in range(num_epochs):
    model.train()
    for batch in train_loader:
        input_ids = batch['input_ids'].to(device)
        attention_mask = batch['attention_mask'].to(device)
        labels = batch['labels'].to(device)

        optimizer.zero_grad()
        outputs = model(input_ids)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()

    # Doğrulama
    model.eval()
    val_loss = 0
    correct = 0
    total = 0
    with torch.no_grad():
        for batch in val_loader:
            input_ids = batch['input_ids'].to(device)
            attention_mask = batch['attention_mask'].to(device)
            labels = batch['labels'].to(device)

            outputs = model(input_ids)
            loss = criterion(outputs, labels)
            val_loss += loss.item()

            _, predicted = torch.max(outputs.data, 1)
            total += labels.size(0)
            correct += (predicted == labels).sum().item()

    print(f'Epoch {epoch+1}/{num_epochs}, '
          f'Train Loss: {loss.item():.4f}, '
          f'Val Loss: {val_loss/len(val_loader):.4f}, '
          f'Val Accuracy: {100 * correct / total:.2f}%')

# Tahmin fonksiyonu
def predict(sentence, keyword):
    model.eval()
    encoding = tokenizer.encode_plus(
        sentence,
        keyword,
        add_special_tokens=True,
        max_length=max_length,
        return_token_type_ids=False,
        padding='max_length',
        truncation=True,
        return_attention_mask=True,
        return_tensors='pt'
    )
    input_ids = encoding['input_ids'].to(device)
    attention_mask = encoding['attention_mask'].to(device)

    with torch.no_grad():
        outputs = model(input_ids)
        _, predicted = torch.max(outputs.data, 1)

    # Sayısal etiketi orijinal etikete dönüştürme
    original_label = label_encoder.inverse_transform(predicted.cpu().numpy())[0]
    return original_label

# Örnek tahmin
sample_sentence = "İnternet, dünya çapında bir iletişim ağıdır."
sample_keyword = "İnternet"
predicted_label = predict(sample_sentence, sample_keyword)
print(f"Tahmin edilen etiket: {predicted_label}")
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM, Trainer, TrainingArguments
import torch



#df = pd.DataFrame(dataset)

# 2. Gerekli kütüphaneleri yükleme
# !pip install transformers pandas numpy scikit-learn torch

# 3. Veri ön işleme
def preprocess(row):
    return f"Sentence: {row['sentence']} Keyword: {row['keyword']}"

df['input_text'] = df.apply(preprocess, axis=1)

# Veri setini eğitim ve test olarak bölme
train_df, test_df = train_test_split(df, test_size=0.2, random_state=42)

# 4. Model seçimi ve eğitimi
model_name = "t5-small"  # Daha büyük modeller için "t5-base" veya "t5-large" kullanabilirsiniz
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSeq2SeqLM.from_pretrained(model_name)


device = torch.device("cuda")
model = model.to(device)

# Veriyi model için hazırlama
class SummarizationDataset(torch.utils.data.Dataset):
    def __init__(self, df, tokenizer, max_input_length, max_target_length):
        self.df = df
        self.tokenizer = tokenizer
        self.max_input_length = max_input_length
        self.max_target_length = max_target_length
    
    def __len__(self):
        return len(self.df)
    
    def __getitem__(self, idx):
        input_text = self.df.iloc[idx]['input_text']
        target_text = self.df.iloc[idx]['label']
        
        inputs = self.tokenizer(input_text, max_length=self.max_input_length, truncation=True, padding='max_length', return_tensors='pt')
        targets = self.tokenizer(target_text, max_length=self.max_target_length, truncation=True, padding='max_length', return_tensors='pt')
        
        input_ids = inputs.input_ids.squeeze()
        attention_mask = inputs.attention_mask.squeeze()
        labels = targets.input_ids.squeeze()
        
        return {
            'input_ids': input_ids,
            'attention_mask': attention_mask,
            'labels': labels
        }

# Veri setlerini oluşturma
train_dataset = SummarizationDataset(train_df, tokenizer, max_input_length=512, max_target_length=128)
test_dataset = SummarizationDataset(test_df, tokenizer, max_input_length=512, max_target_length=128)

# Eğitim argümanlarını ayarlama
training_args = TrainingArguments(
    output_dir='./results',
    num_train_epochs=150,
    per_device_train_batch_size=4,
    per_device_eval_batch_size=4,
    warmup_steps=500,
    weight_decay=0.01,
    logging_dir='./logs',
)

# Eğitici oluşturma ve eğitim
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=train_dataset,
    eval_dataset=test_dataset
)

trainer.train()

# 5. Model değerlendirmesi ve test
def generate_summary(text):
    device = 'cuda' if torch.cuda.is_available() else 'cpu'
    model.to(device)
    inputs = tokenizer(text, return_tensors='pt', max_length=512, truncation=True)
    inputs = {k: v.to(device) for k, v in inputs.items()}
    summary_ids = model.generate(inputs['input_ids'], max_length=128, num_beams=4, early_stopping=True)
    return tokenizer.decode(summary_ids[0], skip_special_tokens=True)
# Test
test_sentence = "CyberWave'in sağladığı yazılım çözümleri etkili ama bazı özelliklerin entegrasyonu zor. Entegrasyon süreçlerini iyileştirirlerse, kullanıcı deneyimi çok daha iyi olabilir."
test_keyword = "CyberWave"
test_input = f"Sentence: {test_sentence} Keyword: {test_keyword}"

generated_summary = generate_summary(test_input)
print(f"Generated Summary: {generated_summary}")
```
